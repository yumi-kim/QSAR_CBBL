```python
'''
## ## 2019. 4.1
## updated by 2019. 5.30

by Yumi Kim CBBL at GNU
Reursive Feature Selection
Given an external estimator that assigns weights to features 
(e.g., the coefficients of a linear model), 
the goal of recursive feature elimination (RFE) is to select features 
by recursively considering smaller and smaller sets of features.
'''

# import Libraries
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.feature_selection import RFE

# Read from csv file
raw_dataset = pd.read_csv('train_x_removing_high_correlation_feature_values_with_y.csv')

if 'smile' in raw_dataset.columns:
    smile_data = raw_dataset['smile'] 
    #print(smile_data)
    raw_dataset = raw_dataset.drop("smile", axis = 1) # Remove "SMILE" column

# Get Data and Target from raw_dataset
y_true_data = []
if 'y_true' in raw_dataset.columns:
    y_true_data = raw_dataset['y_true'] 
    #print(y_true_data)
    raw_dataset = raw_dataset.drop("y_true", axis = 1) # Remove "y_true" column

##### Machine Learning Algorithm
##### Tree-based Feature Selection
X = raw_dataset
X.shape
y = y_true_data

#For Debug
print(X.info())


##### Machine Learning Algorithm
##### Step 5 : Recursive Feature Eliminiation(RFE)
svc = SVC(kernel="linear", C=1)

## n_features_to_select : int or None (default=None)
## The number of features to select. If None, half of the features are selected.
'''
step : int or float, optional (default=1)
If greater than or equal to 1,
then step corresponds to the (integer) number of features to remove at each iteration. 
If within (0.0, 1.0), 
then step corresponds to the percentage (rounded down) of features to remove at each iteration.
'''
selector = RFE(estimator=svc, n_features_to_select=None, step=1)

#selector  = selector.fit(X, y)
sel_data  = selector.fit_transform(X, y)

X_selected_df = pd.DataFrame(sel_data, columns=[raw_dataset.columns[i] for i in range(len(raw_dataset.columns)) if selector.get_support()[i]])

# To write for Univariate Feature Selection(UFS)
print("============================================")
print("After Processing of Recursive Feature Elimination(RFE)")
print("============================================")
print(X_selected_df.info())
X_selected_df['smile']=smile_data # appending for removed "SMILE" column
X_selected_df.to_csv('train_x_recursive_feature_elimination_data.csv', index=False)

```
